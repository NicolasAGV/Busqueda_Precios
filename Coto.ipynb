```python
list_brand = ['Absolut', 'Adler', 'Aguila', 'Aimé', 'Alaris', 'Alcazar', 'Alicante', 'Altamira', 'Amauta', 'Americana', 'Amstel', 'Anclas', 'Andeluna', 'Andes', 
'Angélica', 'Antares', 'Aperol', 'Aquarius', 'Arcor', 'Arla', 'Arrivata', 'Stella Artois', 'Aruma', 'Arytza', 'Atuel', 'Azait', 'Ballantines', 'Ballarini',
'Barraza', 'Bavaria', 'Beldevere', 'Belgian', 'Benjamin', 'Bianchi', 'Bieckert', 'Bierhaus', 'Bimbo', 'Blancaflor', 'Bocatti', 'Bohemian', 'Borghetti',
'Brahma', 'Branca', 'Budweiser', 'Cachafaz', 'Cada Dia', 'Calchaqui', 'Callia', 'Campagnola', 'Campari', 'Canut', 'Casal', 'Casalta', 
'Casancrem', 'Caseritos', 'Catena', 'Cañuelas', 'Cedros', 'Celia', 'Cellier', 'Cepera', 'Chachingo', 'Chango', 'Chia Graal', 'Chocolinas', 
'Cindor', 'Cinzano', 'Clight', 'Cocinero', 'Colavita', 'Conaprole', 'Cormillot', 'Cortijo', 'Coto', 'Cows Pure', 'Cremette', 'Cremix', 
'Croc', 'Crunchy', 'Cuisine & Co Nbe Mp', 'Cunnington', 'Cusine-Co', 'Dada', 'Damonjag', 'Danette', 'Daniels', 'Delfina', 
'Delicia', 'Dodoni', 'Dolca', 'Domenico Renieri', 'Doritos', 'Dos Anclas', 'Doña Aurora', 'Estilo Real', 'Eazy', 'Elementos', 'Emeth',
'Emperador', 'Eneldo', 'Estancia', 'Estelar', 'Exquisita', 'Felices Las Vacas', 'Fabre', 'Famiglia', 'Familia Zuccardi', 'Fanta',
'Fargo', 'Favaloro', 'Favinc', 'Favinco', 'Felicien', 'Felipe', 'Fini', 'Finladia', 'Finlandia', 'Fridays', 'Fritolim', 'Gancia',
'Gandara', 'Gatorade', 'Giallo', 'Giraso', 'Glaciar', 'Glenlivet', 'God Bless You', 'Goose', 'Graffigna', 'Grolsch', 'Guinness', 
'Haroldos', 'Hashi', 'Havana', 'Havarti', 'Heineken', 'Heinz', 'Helles', 'Herrero', 'Hojaldradas', 
'Holanda', 'Huberto', 'Humberto', 'Humos', 'Ignacio', 'Iguana', 'Ile De Fran', 'Ilolay', 'Imperial', 'Jumbo', 'Jumbo Mp', 'Kansas', 
'Kaufman', 'Kelloggs', 'Khune', 'Kitkat', 'Knorr', 'Krachitos', 'Kuhne', 'La Paulina', 'La Quesera', 'La Serenisima', 'La Suerte', 
'La Campagnola', 'La Española', 'La Parmesana', 'La Toscana', 'Landvik', 'Lays', 'Legitimo', 'Lepes', 'Levite', 'Lindaflor', 'Lira', 
'Lucchetti', 'Magnum', 'Maille', 'Malibu', 'Mantecol', 'Manzoni', 'Marca No Identificada', 'Marian Arytza', 'Martini', 'Mayo',
'Mayoliva', 'Mazola', 'Mazzetti', 'Mañanita', 'Melba', 'Melincué', 'Menoyo', 'Mex', 'Milka', 'Milkaut', 'Mirinda', 'Molinos', 
'Monster', 'Multifrua', 'Multifruta', 'Mustard', 'Mutti', 'Naranchelo', 'Natura', 'Nescafé', 'Nesquik', 'Nestlé', 'Neufchatel',
'Nieto', 'Noal', 'Nobles', 'Nonna Pia', 'Norton', 'Not Cheese', 'Notcheese', 'Notcream', 'Notcreamcheese', 'Notmayo', 'Notmayo', 
'Notmilk', 'Novecento', 'Nucete', 'Obrero', 'Oliovita', 'Oreo', 'Orieta', 'Primera Marca', 'Punta Del Agua', 'Padano', 'Paladini', 
'Pampa Gourmet', 'Pampagourmet', 'Pascual', 'Patagonia', 'Pepitos', 'Pepsi', 'Piedras Blancas', 'Pimms', 'Porteña', 'Pringles', 'Pureza',
'Pyros', 'Qu', 'Quesitos', 'Quilmes', 'Richard Lac', 'Ramazzotti', 'Recetas De Entonces', 'Riojana', 'Robino', 'Roble', 'Rocklets', 
'Rosenborg', 'Rutini', 'Santa Rosa', 'Salentein', 'San Giorgio', 'Sancor', 'Santa Clara', 'Savora', 'Schneider', 'Schweppes', 
'Seltzer', 'Serenísima', 'Serrano', 'Sheridan', 'Silvia', 'Sin Marca', 'Singleton', 'Skyy', 'Smirnoff', 'Spirito', 'Sprite', 'Spritz', 
'Sriracha', 'Strega', 'Swift', 'Tabassher', 'Talisman', 'Tang', 'Tau Delta', 'Tentaciones', 'Teresa', 'Terma', 'The Not Co',
'Toddy', 'Tonadita', 'Trapiche', 'Tregar', 'Trivento', 'Tropical', 'Tropicales', 'Tía', 'Vacalin', 'Valderrobles', 'Valentin', 
'Veronica', 'Via', 'Villavicencio', 'Vincezo', 'Virginia', 'Waldbier', 'Walker', 'Wapi', 'Warsteiner', 'Wings', 'Yancanello',
'Yogurisimo', 'Zanetti', 'Zucaritas', 'Zuelo', 'Corona', 'Miller', 'Sol', 'Cusqueña', 'Temple', 'Rabieta', 'Siete Chola', 'Salta', 
'Pampa Brewing Co', 'Oranjeboom', 'Blue Moon', 'Kunstmann', 'Hertog Jan', 'Michelob', 'Lit','Cabrales', 'Nescafé', 'Starbucks', 'Bonafide',
'Lavazza', 'Morenita', 'Martinez', 'Arlistan', 'El Pocillo', 'Coffee Mate', 'Macchiato', 'Maltife', 'Higienol', 'Campanita', 'Felpita', 'Elite', 'Cristal Del Lago', 
'Bariloche', 'Kinder', 'Cofler', 'Block', 'Shot', 'Cadbury', 'Georgalos', 'Vizzio', '1891 Neugebauer', 'Tokke', 'Kitkat', 'Belgium', 'Classic Wheels', '1891', 
'The Happy Snack', 'Pepitos', 'Fantoche', 'Marroc', 'Shot', 'Yapa', 'Sugus', 'Jorgito', 'Fulbito', 'Tofi', 'Menthoplus', 'Lenguetazo', 'Chocoarroz', 'Halls', 'Terrabusi', 
'Cachafaz', 'Topline', 'Leger', 'Lheritier', 'Cofler', 'Jorgelín', 'Chocolatin', 'Jorgelin', 'Neugebauer', 'Beldent', 'Luján', 'Mogul', 'Cebritas', 'Minitorta', 'Marengo', 
'Milka', 'Bagley', 'Lipo', 'Billiken', 'Bon O Bon', 'Alfa Pampa', 'Sin Culpa', 'Palitos De La Selva', 'Pico Dulce', 'Flynn Paff', 'Push Pop', 'Ring Pop', 'Butter Toffees',
'Bull Dog', 'Tic Tac', ' Bazooka', 'Santa Maria', 'La Campagnola', 'Aleluya', 'Buffalo', '', '', '', '', '', '', '', '', '', '', '', '', '', '', 
]

dict_brands_wrong = {'La Paulin':'La Paulina', 'Santa Ros':'Santa Rosa', 'Arri':'Arrivata', 'Punta Del A':'Punta Del Agua', 'Melincue':'Melincué', 'Piedras Bla':'Piedras Blancas', 
                     'Stella Arto':'Stella Artois', 'Abstemia':'Peñón Del Aguila', 'Nescafé®':'Nescafé', 'Starbucks®':'Starbucks', 'Nescafe':'Nescafé', 'Martin':'Martinez',
                      'Cristal Del':'Cristal Del Lago', 'Kitkat®':'Kitkat', 'Flynnies':'Georgalos', "Mr.Pop S": "Mr.Pop'S", "Mr.Pop'S": "Mr.Pop'S", 'Mentho Plus':'Menthoplus',
                    'Chupetón Cito': 'Lheritier', 'Butter Toff':'Butter Toffees', 'Wamis':'Lipo', 'Mr Floky':'Marengo', 'Bola Loca': 'Lheritier', '':'', '':'', '':'', '':'',
                    "Hellmann'S":"Hellmann's", "HELLMANNS":"Hellmann's", 'BC':'La Campagnola', 'Bc':'La Campagnola', '':''
                        
                    }
                 
list_category = ['Absenta', 'Aceiitunas', 'Aceite', 'Aceitnas', 'Aceituna', 'Aceitunas', 'Aceto', 'Actimel', 'Activia', 'Aderezo', 'Ades', 'Agua', 'Aji', 'Ajo', 'Alfajor',
                 'Alfajores', 'Alim', 'Alimento', 'Almidon', 'Amargo', 'Anchoas', 'Antipasto', 'Aperitivo', 'Apricot', 'Arroz', 'Arvejas', 'Atun', 'Avena', 'Azucar', 'Azúcar',
                 'Babaganush', 'Balsámico', 'Bananita', 'Barbacoa', 'Bbq', 'Bebida', 'Bicarbonato', 'Bizcochuelo', 'Bocconcino', 'Bolsa', 'Bolsas', 'Bombones', 'Bondiola',
                 'Brownies', 'Budín', 'Burrata', 'Cachaza', 'Cafe', 'Café', 'Caldo', 'Cappelletti', 'Centro', 'Cerveza', 'Cheddar', 'Chili', 'Chimichurri', 'Choclo',
                 'Chocolatada', 'Chocolate', 'Chorizo', 'Chucrut', 'Chutney', 'Coca-cola', 'Cognac', 'Conito', 'Cordero', 'Crema', 'Criollo', 'Cuatro', 'Curry', 
                 'Deshidrat', 'Detox', 'Dip', 'Don', 'Dulce', 'Duraznos', 'Energizante', 'Esencia', 'Esparragos', 'Fernet', 'Fiambre', 'Fideos', 'Flan', 'Fondue',
                 'Fresa', 'Frizante', 'Fuet', 'Galletita', 'Galletitas', 'Garbanzos', 'Gaseosa', 'Gelatina', 'Gift', 'Gin', 'Girasol','Golf', 'Gomitas', 'Granadina',
                 'Granos', 'Grappa', 'Grasa', 'Growler', 'Hard', 'Harina', 'Hielo', 'Honey', 'Hummus', 'Isotónica', 'Jamon', 'Jamón', 'Jugo', 'Ketchup', 'Kiwi',
                 'Kombucha','Lacteo', 'Leber', 'Leberwurst', 'Leche', 'Lentejas', 'Levadura', 'Licor', 'Limonada', 'Lomo', 'Longaniza', 'Maille', 'Manteca',
                 'Maní', 'Margarina', 'Marinade', 'Masa', 'Mayonesa', 'Maíz', 'Membrillo', 'Milk', 'Mingo', 'Mix', 'Mortadela', 'Mostaza', 'Mousse',
                 'Mozzarella', 'Noquis', 'Nutella', 'Oblea', 'Oliva', 'Oregano', 'Orégano', 'Pack', 'Paleta', 'Palitos', 'Palmitos', 'Pampagourmet', 'Pan', 'Panceta',
                 'Panzotti', 'Papas', 'Pascualina', 'Pasta', 'Pate', 'Pechuga', 'Pepino', 'Pepinos', 'Pesto', 'Picada', 'Picadillo', 'Pickles', 'Pimenton',
                 'Pisco', 'Polpetta', 'Pomelada', 'Postre', 'Provenzal', 'Provoleta', 'Punt', 'Pure', 'Puré', 'Q.blancrem', 'Q.procesado', 'Qso', 'Quesitos', 'Queso',
                 'Rallado', 'Ramen', 'Ranch', 'Ravioles', 'Ravioli', 'Raviolitos', 'Raviolones', 'Repollo', 'Ricota', 'Ricotta', 'Risotto', 'Ron','Rotisera', 'Saborizador', 
                 'Sal', 'Salame', 'Salamin', 'Salamín', 'Salchichas', 'Salchichón', 'Salmón', 'Sals', 'Salsa', 'Salsas', 'Rotisera', 'Saborizador', 'Sidra', 'Snack', 'Soda',
                 'Sopa', 'Sopresatta', 'Sorrentino', 'Spianata', 'Spianataa', 'Spianatta', 'Spicy', 'Sésamo', 'T-qso', 'Tabla', 'Tapa', 'Tapas', 'Teriyaki', 'Tita', 'Tomates', 
                 'Tortellini', 'Tortelloni', 'Tortillas', 'Tradicional', 'Trucha', 'Turrón', 'Vacas', 'Vainillas', 'Vermouth', 'Vin', 'Vinagre', 'Vinagres', 'Vino', 'Vodka',
                 'Whisky', 'Yerba', 'Yog', 'Yogur', 'Yogurt', 'Ñoqui', 'Ñoquis', 'Ñoquis']

def sort_and_capitalize (list_1):
    list_1.sort()
    list_2 = []
    for a in list_1:
        list_2.append(a.title())

    return list_2

def search_brand_wrong_words (text, dict_brand):               # Busca marcas incorrectas
    busqueda = 'Marca no identificada'
    dict_1 = dict_brand

    for key, value in dict_1.items():
        var1 = re.search(key, text,re.IGNORECASE)
        if var1 != None:
            busqueda = value

    return busqueda

def search_brand_many_words (text, list_brand):                 # Busca marca de varias palabras correctas
    list_1 = [x for x in list_brand if len(x.split())>1]
    busqueda = 'Marca no identificada'
    flag = False
    k = 0

    while flag == False and k < len(list_1):
        var1 = re.search(list_1[k], text,re.IGNORECASE)
        k += 1
        if var1 != None:
            flag = True
            busqueda = var1[0].title()

    if busqueda == 'Marca no identificada':
        busqueda = search_brand_wrong_words(text, dict_brands_wrong)

    return busqueda

def product_brand_category(var_1, list_brand):                # Buca marca de una palabra correcta
    var_2_split = var_1.split()
    list_1 = [x for x in list_brand if len(x.split())==1]
    marca = 'Marca no identificada'
    categoria = 'Categoria no identificada' 

    for item in var_2_split:
        if item.capitalize() in list_1:
            marca = item.capitalize()
        # else:
        #     marca = buscador_de_marcas(item.capitalize, dict_marcas_erradas)
        if item.capitalize() in list_category:
            categoria = item
               
    if marca == 'Marca no identificada':
        marca = search_brand_many_words(var_1, list_brand )

    return marca, categoria 



# **********----------**********----------**********----------**********----------**********----------**********----------**********----------
# DISCOUNT SECTION

def accept_discount (var):
    var_1 = var.find("div", {"class": "promo_bottom"}).find("img")
    if var_1 == None:
        return True
    else:
        return False

def discount_any(var):                                # comprueba si existe algun tipo de descuento buscando 'price_regular_precio'
    try:
        return var.find("span", {"class": "price_regular_precio"}).text
    except:
        return 'no hay descuento ni promo'
        
def exist_desc_llevandoN (var):
    try:
        return var.find("div", {"class": "desc-llevandoN"}).text
    except:
        return 'no hay desc_llevandoN'
    
def amount_of_discount_container (var):                                 # Cantidad de 'discount_containers'
    return len(var.findAll("span", {"class": "text_price_discount"}))

def get_price_discount (var):
    try:
        return var.find("span", {"class": "price_discount"}).text

    except:
        return var.find("span", {"class": "price_discount_gde"}).text

def discount_legend (var):  # Prueba conseguir los datos de 'desc-llevandoN' y containers 1 y 2. Si hay mas, fue. Tambien cantidad de containers
    list_3 = []

    try:
        list_3.append(var.find("div", {"class": "desc-llevandoN"}).text.replace('\n', '').replace('\t', '').strip() )
    except:
        pass

    try:
        list_3.append(var.find("div", {"class": "first_price_discount_container"}).text.replace('\n', '').replace('\t', '').strip() )
    except:
        pass

    try:
        list_3.append(var.find("div", {"class": "second_price_discount_container"}).text.replace('\n', '').replace('\t', '').strip() )
    except:
        pass

    try:
        list_3.append(amount_of_discount_container(var))
    except:
        pass
       
    return list_3

def discount_split (var):
    discount_direct, discount_pXq, discount_ComCot = '-', '-', '-'
    list_var = discount_legend(var)
    if list_var[-1] == 1:
        if 'Llevando' in list_var [0]:
            discount_pXq = float(list_var[1].split('$')[1][:-3])
        elif 'Dto' in list_var [0]:
            discount_direct = float(list_var[0].split('$')[1])
    if list_var[-1] == 2:
        if 'Dto' in list_var [0]:
            discount_direct = float(list_var[0].split('$')[1])
        discount_ComCot = float(list_var[1].split('$')[1])

    return discount_direct, discount_pXq, discount_ComCot

def discount_porcentage (value, porcentage):
    return (value/100) * (100-porcentage)

def lowest_price (PPPF, discount_direct, discount_pXq, discount_ComCot):
    list1 = []
    for a in [PPPF, discount_direct, discount_pXq, discount_ComCot]:
        list1.append(a)
    
    return pd.Series([x for x in list1 if x != '-']).min()

def discount_special (price, condition):              # Indica dia, cantidad y condicion de descuento por condition (retired, card, etc)
    cond_var = condition.title()
    legend = 'No hay descuento aplicable'
    if cond_var == 'Jubilado' or cond_var == 'Pensionado':
        amt_disc = 15
        legend = str(amt_disc) + '% '  + 'Lunes y Jueves'
        price = discount_porcentage(price, amt_disc)
    elif cond_var == 'Icbc':
        amt_disc = 20
        legend = str(amt_disc) + '% '  + 'Jueves'
        price = discount_porcentage(price, amt_disc)
    elif cond_var == 'Tci':
        amt_disc = 20
        legend = str(amt_disc) + '% '  + 'Lunes'
        price = discount_porcentage(price, amt_disc)
    elif cond_var == 'Credito':
        amt_disc = 15
        legend = str(amt_disc) + '% '  + 'Del Viernes 15/3 al Domingo 17/3'
        price = discount_porcentage(price, amt_disc)    

    return cond_var, legend, price

def re_format_df_for_discount_special (df, condition, disc_rand_leg):
    df_new = df.copy()
    df_new.reset_index(drop=True, inplace=True)
    list_legend = []
    list_price_posted = []
    list_price_main_unit = []
    list_condition = []

    for a in range(len(df_new)):
        if df_new.loc[a, disc_rand_leg][0] == 0 and df_new.loc[a, 'Acept_Disc']:
            list_legend.append(discount_special(df_new.loc[a, 'P_Price'], condition)[1])
            list_condition.append(discount_special(df_new.loc[a, 'P_Price'], condition)[0])
            list_price_posted.append(discount_special(df_new.loc[a, 'P_Price'], condition)[2])
            list_price_main_unit.append(discount_special(df_new.loc[a, 'CMU_Price'], condition)[2])
            df_new.loc[a, 'Lowest_Price_Prod'] = list_price_posted[-1]
            df_new.loc[a, 'LP_MainUnit'] = list_price_main_unit[-1]

        else:
            list_legend.append('-')
            list_condition.append('-')
            list_price_posted.append('-')
            list_price_main_unit.append('-')

        
    df_new.insert(16, 'DC_MainUnitPrice', pd.Series(list_price_main_unit))
    df_new.insert(16, 'DC_Price', pd.Series(list_price_posted))
    df_new.insert(16, 'DC_Legend', pd.Series(list_legend))
    df_new.insert(16, 'Disc_Cond', pd.Series(list_condition))

    return df_new


# **********----------**********----------**********----------**********----------**********----------**********----------**********----------
# VARIOUS

def product_category(var_1):                

    var_2_split = var_1.split()
    categoria = 'Categoria no identificada' 

    for item in var_2_split:
        if item.capitalize() in master_list_supermarket.list_category:
            marca = item
        else:
            marca = buscador_de_marcas(item.capitalize, dict_marcas_erradas)

        if item.capitalize() in list_category:
            categoria = item
               
    return marca, categoria  

def product_description(var):       # devuelve descripcion del producto
    return var.find("div", {"class": "descrip_full"}).text

def post_compare_legend(var):
    asd =  (' ').join(str(var.find("span", {"class": "unit"}).text).split())
    if len(asd) > 3:
        return asd
    else:
        return '-'

def post_compare_price(var):
    return re.search("[0-9\\$,.]+", var.split(':')[1])[0]

def post_compare_amt(var):
    return re.search("[0-9]+", var.split(':')[0])[0]

def post_compare_unit(var):
    return sh.detect_unit_published_prueba(var.split(':')[0])[1]

def lowest_price_per_main_unit(precio_normal, precio_mas_bajo, precio_por_main_unit):
    if precio_por_main_unit != '-':
        return float(round((precio_mas_bajo / precio_normal) * precio_por_main_unit, 1))
    else:
        return '-'

def compare(post_compare_amt, post_compare_unit, post_compare_price_float):  

    unit_M = 'unidad no identificada'   

    price_PMUM = 000                                       
    if post_compare_unit in ['kilogramos']:  
        price_PMUM =  post_compare_price_float 
        unit_M = 'kilogramos'

    elif post_compare_unit in ['litros']:  
        price_PMUM =  post_compare_price_float 
        unit_M = 'litros'
    
    elif post_compare_unit in ['m²']: 
        price_PMUM =  post_compare_price_float            
        unit_M = 'm²'

    elif post_compare_unit in ['unidades']: 
        price_PMUM =  post_compare_price_float            
        unit_M = 'unidades'

    elif post_compare_unit in ['gramos']:              
        price_PMUM = (post_compare_price_float / float(post_compare_amt )) * 1000
        unit_M = 'kilogramos'

    elif post_compare_unit in ['mililitros']:            
        price_PMUM = (post_compare_price_float / float(post_compare_amt )) * 1000
        unit_M = 'litros'    

    return round(float(price_PMUM), 1), unit_M 

def post_compare_with_no_legend (var, price_posted):
    amount = 0000
    unit = 'unidad no identificada'
    price = 0000
    var2 =  re.search("(kgm|kg)", var,re.IGNORECASE)
    if var2 != None:
        unit = 'kilogramos'
        amount = 1
        price = price_posted   
        
    return amount, unit, price

def product_link(var):       # devuelve descripcion del producto
    return 'https://www.cotodigital3.com.ar' + var.find("a").attrs['href']
    
def search_contain (df, key_word):
    return df[df['Post_Description'].str.contains(key_word, case=False)].sort_values(by='LP_MainUnit')

def search_contain_not (df, key_word):
    return df[~df['Post_Description'].str.contains(key_word, case=False)].sort_values(by='LP_MainUnit')

# **********----------**********----------**********----------**********----------**********----------**********----------**********----------
# FROM LINK TO DF

dict_coto_almacen_parte_1d5 = { 
'Alfajores':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-golosinas-alfajores/_/N-1njwjm5',              # Golosinas 
# 'Chocolates':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-golosinas-chocolates/_/N-uiml5b',
# 'Caramelos y chupetines':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-golosinas-caramelos-y-chupetines/_/N-1xkf1n',
# 'Chicles y pastillas':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-golosinas-chicles-y-pastillas/_/N-axoedr',    
# 'Mayonesas':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-aderezos-y-salsas-mayonesas/_/N-r3vqlx',       # Aderezos y salsas 
# 'Mostaza':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-aderezos-y-salsas-mostazas/_/N-4dl2yh',
# 'Ketchup':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-aderezos-y-salsas-ketchup/_/N-aq618b',
# 'Salsas golf':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-aderezos-y-salsas-salsas-golf/_/N-faua8f',
# 'Mermeladas':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-mermeladas-y-dulces-mermeladas/_/N-va3r6f',   # Mermeladas y dulces
# 'Dulce':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-mermeladas-y-dulces-dulce/_/N-xa8giw',
# 'Miel':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-mermeladas-y-dulces-miel/_/N-10v5iow',
#                                                                                                                                     # 'Jalea':'',
# 'Especias':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-especias-especias/_/N-1arvz2p',                 # Especias
# 'Semillas':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-especias-semillas/_/N-1wi2j1a',
# 'Leche en polvo':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-leche-en-polvo/_/N-1yw5bwj'              # Leche en polvo 
}

dict_coto_almacen_parte_2d5 = { 
'Galletitas':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-panaderia-galletitas/_/N-10z239c',                  # Panaderia
'Panaderia propia':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-panaderia-panaderia-propia/_/N-1s5i92o', 
'Panificados':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-panaderia-panificados/_/N-1sv7ob1', 
'Galletas, tostadas y grisines':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-panaderia-galletas-tostadas-y-grisines/_/N-39hspl', 
'Cafe':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-infusiones-caf%C3%A9/_/N-1cgicr5',                # 'Infusiones'
'Mate':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-infusiones-mate/_/N-vra9dh',
'Mate Cocido':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-infusiones-mate-cocido/_/N-12sv9y0',
'Te':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-infusiones-t%C3%A9/_/N-vtu7ep',      
'Puré de tomate':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-salsas-y-pur%C3%A9-de-tomate-pur%C3%A9-de-tomate/_/N-1eekjet',    #'Salsas y puré de tomate'
'Pulpa de tomate':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-salsas-y-pur%C3%A9-de-tomate-pulpa-de-tomate/_/N-kf0bve',
'Extracto tomates':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-salsas-y-pur%C3%A9-de-tomate-extracto-tomates/_/N-15iy0o5',
'Salsas preparadas':'http://api.cotodigital.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-salsas-y-pur%C3%A9-de-tomate-salsas-preparadas/_/N-1gnfvbp',
'Pasta lista y rellena':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-pasta-seca-lista-y-rellenas-pasta-lista-y-rellena/_/N-3w7irq', # Pasta seca, lista y rellenas
'Pastas secas':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-pasta-seca-lista-y-rellenas-pastas-secas/_/N-f258hq', 
'Suplementos dietarios':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-suplementos-dietarios/_/N-qe3p7f'  # Suplementos dietarios
}

dict_coto_almacen_parte_3d5 = { 
'Papas fritas':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-snacks-papas-fritas/_/N-qs05qq',                               # Snacks
'Bastones de maiz':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-snacks-bastones-de-maiz/_/N-8fgqyh', 
'Mani':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-snacks-mani/_/N-l50on3', 
'Palitos':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-snacks-palitos/_/N-1ak1mib', 
'Conservas de carne':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-conservas-conservas-de-carne/_/N-1aa25uw',                  # Conservas
'Conservas de frutas':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-conservas-conservas-de-frutas/_/N-13arqj7', 
'Conservas de pescado':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-conservas-conservas-de-pescado/_/N-17wcrsq', 
'Conservas vegetal':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-conservas-conservas-vegetal/_/N-1eywm6f', 
'Aceites':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-aceites-y-condimentos-aceites/_/N-16r0nc0',                            # Aceites y condimentos 
'Condimentos':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-aceites-y-condimentos-condimentos/_/N-1f9ytwc', 
'Vinagres':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-aceites-y-condimentos-vinagres/_/N-1a5jpj4', 
'Premezcla postres':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-polvo-para-postres-y-reposteria-premezcla-postres/_/N-1pm2ots', # Polvo para postres y reposteria 
'Repostería':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-polvo-para-postres-y-reposteria-reposter%C3%ADa/_/N-tmyd8',
'Productos orgánicos':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-productos-org%C3%A1nicos/_/N-b9ra0a',                      # Productos orgánicos
}

dict_coto_almacen_parte_4d5 = {
'Cereales con azúcar':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-cereales-cereales-con-az%C3%BAcar/_/N-1jlp6ab', # Cereales
'Cereales infantiles':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-cereales-cereales-infantiles/_/N-146vmzh',
'Break saludables':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-cereales-break-saludables/_/N-1ktgydt',
'Barras de arroz':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-cereales-barras-de-arroz/_/N-1wmpv6n',
'Aceitunas':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-encurtidos-aceitunas/_/N-kikopr', # Encurtidos
'Ají':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-encurtidos-aj%C3%AD/_/N-18099ns',
'Alcaparras':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-encurtidos-alcaparras/_/N-158qwuq',
'Alcaucil':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-encurtidos-alcaucil/_/N-zjqlo5',
'Alimentos para bebé':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-alimento-de-beb%C3%A9s-y-ni%C3%B1os-alimentos-para-beb%C3%A9/_/N-hirzwy', # Alimento de bebés y niños
'Infusiones infantiles':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-alimento-de-beb%C3%A9s-y-ni%C3%B1os-infusiones-infantiles/_/N-7h3to3',
'Mezcla lista':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-sopas-caldos-pur%C3%A9-y-saborizantes-mezcla-lista/_/N-3t1dg7', # Sopas, caldos, puré y saborizantes
'Sopas y saborizantes':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-sopas-caldos-pur%C3%A9-y-saborizantes-sopas-y-saborizantes/_/N-1v2h1e8',
'Huevos de pascua grandes':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-huevos-de-pascua-huevos-de-pascua-grandes/_/N-bpo2vl', # Huevos de pascua
'Huevos de pascua medianos':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-huevos-de-pascua-huevos-de-pascua-medianos/_/N-12ets07',
'Huevos de pascua pequeños':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-huevos-de-pascua-huevos-de-pascua-peque%C3%B1os/_/N-cnrvrp',
'Conejos y figuras de chocolate':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-huevos-de-pascua-conejos-y-figuras-de-chocolate/_/N-1mg249u',
}

dict_coto_almacen_parte_5d5 = {
'Azúcar':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-endulzantes-az%C3%BAcar/_/N-1w1x9xa', # Endulzantes
'Edulcorante':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-endulzantes-edulcorante/_/N-sow8s0',
'Harina de arroz':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-harinas-harina-de-arroz/_/N-6zd215', # Harinas
'Harina de trigo':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-harinas-harina-de-trigo/_/N-14jn5rm',
'Harina de maíz':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-harinas-harina-de-ma%C3%ADz/_/N-i6ajfm',
'Premezcla':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-harinas-premezcla/_/N-1jtpcin',
'Arroz':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-arroz-y-legumbres-arroz/_/N-149fj7g', # Arroz y legumbres
'Legumbres':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-arroz-y-legumbres-legumbres/_/N-4trlda',
'Pan rallado':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-rebozador-y-pan-rallado-pan-rallado/_/N-o13ttg', # Rebozador y pan rallado
'Rebozador':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-almac%C3%A9n-rebozador-y-pan-rallado-rebozador/_/N-x0s2im',
}

dict_coto_bebidas = {
# 'Aguas':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-sin-alcohol-aguas/_/N-17xtqou', #Bebidas sin alcohol
# 'Gaseosas':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-sin-alcohol-gaseosas/_/N-n4l4r5',
# 'Jugos':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-sin-alcohol-jugos/_/N-11la5tu',
# 'Amargos':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-sin-alcohol-amargos/_/N-1mf4idu',
# 'Aperitivos':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-con-alcohol-aperitivos/_/N-1lbig2w', # Bebidas con alcohol
'Cerveza':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-con-alcohol-cerveza/_/N-137sk0z',
# 'Champagne y espumantes':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-con-alcohol-champagne-y-espumantes/_/N-qxxhmm',
# 'Vinos':'https://www.cotodigital3.com.ar/sitios/cdigi/browse/catalogo-bebidas-bebidas-con-alcohol-vinos/_/N-uqiqtm'
}













def link_to_df_categories (dict_with_category):     # Recibe por medio del dict la categoria principal y devuelve dfs concatenados de las subs categorias
    pirulo = dict_with_category
    list_df = []
    for key, value in pirulo.items():
        print(key)
        BS_objt = link_to_BsObj(value)
        list_df.append(list_BsObj_to_df(BS_objt, key))
        time.sleep(5 * np.random.random())
        print('Duerme 5')

    df_new = list_df[0].copy()
    a=1
    while a < len(list_df):
        df_new = pd.concat([df_new, list_df[a]]).copy()
        a += 1
        
    df_new_2 = df_new.reset_index(drop=True)     

    return df_new_2

def list_link_to_list_BsObj (list_link):    # Con lista de links devuelve lista de BS OBJ a nivel pagina (con el frame con productos). Ejemplo:
    list_soup_pages = []                    #  list = [BsObj_1.1, BsObj_1.2, BsObj_2.1, BsObj_2.2] 

    for link in list_link:
        list_soup_pages += link_to_BsObj(link)
        
    return list_soup_pages   

def link_to_BsObj (link):           # Devuelve BS_OBJ de productos hasta el fin de las paginas de dicho link
    top = True
    list_var = []
    count = 0    
    page = 1
    link_first_part = link          # Link de pagina numero 1 de la categoria
    link_addon_1 = '?Nf=product.startDate%7CLTEQ+1.7082144E12%7C%7Cproduct.endDate%7CGTEQ+1.7082144E12&No='
    link_addon_2 = '&Nr=AND%28product.language%3Aespa%C3%B1ol%2Cproduct.sDisp_200%3A1004%2COR%28product.siteId%3ACotoDigital%29%29&Nrpp=12'

    while top == True:
        response = requests.get(link)
        soup = BS(response.content)
        soup_2 = soup.findAll("li", {"class": "clearfix"}) 
        print(f'Pagina {page} seccion {count}')
        
        if len(soup_2) > 0:                      # Chequea que encontro algun 'li'...
            list_var += soup_2
            time.sleep(3 * np.random.random())      # hace tiempo    
            print('Duerme 3')            
            count += 12
            page += 1                                # usa dos mitades de links con un numero mutiplo de 12 en el medio
            link =  link_first_part + link_addon_1 + str(count) + link_addon_2

        else:
            top = False   

    return list_var  

def list_BsObj_to_df (list_BsObj_products, category):         # input: 'lista de BS OBJ de productos' o 'str bajado'. Devuelve DF
    list_var_main_dict = []

    if type(list_BsObj_products) == str:                # En el caso que el input sea un txt leido desde 'read'                          
        soup = BS(list_BsObj_products, 'html.parser')
        # Al ser un gran BS OBJ, busco directamente
        list_BsObj_products_2 = soup.findAll("li", {"class": "clearfix"})# Encuentra la cabeza de cada producto

        for BsObj_producto in list_BsObj_products_2:                # Itera sobre los BS OBJ de los productos
            list_var_main_dict.append(main_dict_function(BsObj_producto, category))

    else:                                       # En el caso que venta de una lista de BS OBJ
        try:
            for BsObj_producto in list_BsObj_products:                    # Itera sobre los BS OBJ de los productos
                list_var_main_dict.append(main_dict_function(BsObj_producto, category))

        except:
            print('No se encontro el BS OBJ del producto')

    return pd.DataFrame(list_var_main_dict).reset_index(drop=True) 

def main_dict_function(var, category):    
    main_dict = {}
    
    # Informacion obtenida del posteo del producto (descripcion, marca, cantidad, unidad, precio)
    main_dict['P_Category'] = category
    main_dict['Post_Description'] = product_description(var)                                   # Descripcion
    # print(main_dict['P_description'])
    main_dict['P_Brand'] = product_brand_category(main_dict['Post_Description'], list_brand)[0]   # "'post_product_brand'"          # Coto no categoriza las marcas
    
    main_dict['P_Amt'] = sh.detect_unit_published_prueba(main_dict['Post_Description'])[0]   # Cantidad publicada  "Post Product Amount"
    main_dict['P_Unit'] = sh.detect_unit_published_prueba(main_dict['Post_Description'])[1]   # Unidad publicada   "post_product_unit" 
    main_dict['P_Price'] = sh.pure_float(sh.post_product_price(var, 'coto'))           # Precio publicado (Math)  "post_product_price_float"

    # Informacion obtenida de la leyenda de comparacion (si existe) (leyenda, cantidad, unidad, precio)
    main_dict['Compare_Legend'] = post_compare_legend(var)                             # Leyenda de precio unidad comparada
    if main_dict['Compare_Legend'] != '-':                                             # A veces no hay leyenda de comparacion (quesos enteros)
        main_dict['C_Amt'] = post_compare_amt(main_dict['Compare_Legend'])                        # Cantidad de comparacion publicada   "post_compare_amt"  (para calcular un 'main_unit_compare' en el caso que que tengas 'sub unidades' )
        main_dict['C_Unit'] = post_compare_unit(main_dict['Compare_Legend'])                         # Unidad de comparacion publicada  "post_compare_unit"
        main_dict['C_Price'] = sh.pure_float(post_compare_price(main_dict['Compare_Legend']))      # Precio de comparacion publicada (Math)  "post_compare_price_float"
    else:                                   
        main_dict['C_Amt'] = post_compare_with_no_legend(main_dict['Post_Description'], main_dict['P_Price'])[0]
        main_dict['C_Unit'] = post_compare_with_no_legend(main_dict['Post_Description'], main_dict['P_Price'])[1]
        main_dict['C_Price'] = post_compare_with_no_legend(main_dict['Post_Description'], main_dict['P_Price'])[2]

    # Calcula la unidad master para compara a traves de los productos (kg, litro, etc) y el precio del mismo
    main_dict['CalcMainUnit_Unit'] = compare(main_dict['C_Amt'], main_dict['C_Unit'], main_dict['C_Price'])[1]    # Unidad principal para comparar, "calculated_main_unit_compare"
    main_dict['CMU_Price'] = compare(main_dict['C_Amt'], main_dict['C_Unit'], main_dict['C_Price'])[0]       # Precio por unidad principal (Math) 'calculated_main_unit_price_float'
 
    # Calcula y aplica los distintos descuentos (a dedo, por llevar varios, Comunidad Coto)
    main_dict['Disc_Rand_Leg'] = discount_legend(var)        # Leyenda de descuento aleatorios 
    main_dict['DR_Direct'] = discount_split(var)[0]                 # Descuento directo sin condiciones
    main_dict['DR_pXq'] = discount_split(var)[1]                    # Descuento llevando P x Q
    main_dict['DR_ComCot'] = discount_split(var)[2]                 # Descuento por Comunidad Coto NO general

    # Indica el precio mas bajo por producto y por unidad principal teniendo en cuenta los descuentos
    main_dict['Lowest_Price_Prod'] = lowest_price (main_dict['P_Price'], main_dict['DR_Direct'], main_dict['DR_pXq'], main_dict['DR_ComCot'])
    main_dict['LP_MainUnit'] = lowest_price_per_main_unit(main_dict['P_Price'], main_dict['Lowest_Price_Prod'], main_dict['C_Price'])

    main_dict['Place'] = 'Coto'

    main_dict['Acept_Disc'] = accept_discount(var)


    main_dict['Link'] = product_link(var)
    
    return main_dict
```
